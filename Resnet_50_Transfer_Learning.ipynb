```
import os, shutil
folders = os.listdir("Images") #Accessed the folder
print(folders)
```

```
import numpy as np
import os
from keras.applications.resnet50 import ResNet50 #Will use this pre trained model on the imagenet
from keras.preprocessing import image #Will use it to process images
from keras.optimizers import Adam 
from keras.layers import * #Will import all layers
from keras.models import Model
from keras.utils import np_utils
import random

import matplotlib.pyplot as plt
```

```
image_data = [] #Will store images
labels = [] #The labels of the images

label_dict = {
    "cats" : 0,
    "dogs" : 1,
    "horses" : 2,
    "humans" : 3
}

for ix in folders: #Iterate over each folder
    path = os.path.join("Images/"+ix) #Got the right folder name
    img_data = os.listdir(path)  #The complete path represents one folder
    for im in img_data:
        img = image.load_img(os.path.join(path,im),target_size=(224,224)) #Taking every image in that folder and making complete path and storing image as 224*224
        img_array = image.img_to_array(img) #Converting each image into an array
        image_data.append(img_array) #Storing that array in image_data list
        labels.append(label_dict[ix]) #Storing the labels

print(len(image_data), len(labels))
```

```
combined = list(zip(image_data, labels)) #Combined image_data and labels together
random.shuffle(combined) #Then we shuffled the data

image_data[:], labels[:] = zip(*combined) #Now taking image features and labels separately 
```

```
X_train = np.array(image_data) #converting in nparray
Y_train = np.array(labels)
Y_train = np_utils.to_categorical(Y_train) #Converting Y in to hot code encoder

print(X_train.shape,Y_train.shape)
#The shapes of the two arrays we have
```

```
model = ResNet50(include_top=False,weights='imagenet',input_shape=(224,224,3)) #Don't need top so put it false
#Weights of the model trained on imagenet will be used
```

```
model.summary()
```

```
av = GlobalAveragePooling2D()(model.output) #This is gap layer taking output of model as input
fc1 = Dense(256,activation='relu')(av) #Then output of gap layer serves as input to fully connected layer
d1 = Dropout(0.5)(fc1) #For saving from overfitting
fc2 = Dense(4,activation='softmax')(d1) #Used dense layer of four because we have four classes and using softmax function

model_new = Model(inputs=model.input, outputs= fc2)
model_new.summary()
```

```
adam = Adam(lr=0.00003) #Changing the learning rate which we will apply for just few layers because it is impractical for every layer
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy']) 
```

```

for ix in range(len(model_new.layers)):
    print(ix, model_new.layers[ix])
#Including our layers there are 178 layers
```

```
for ix in range(171):
    model_new.layers[ix].trainable = False #Put first 171 layers as untrainable, so it won't train them
    
model_new.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
model_new.summary()
#I can see that the number of trainable parameters has reduced
```

```
hist = model_new.fit(X_train,Y_train,
                    shuffle = True,
                    batch_size = 16,
                    epochs = 8,
                    validation_split=0.10
                    )
```

```
print(hist.history['acc'][-1]) #accuracy
print(hist.history['val_acc'][-1]) #ValidationAccuracy
#Should have stopped at seven epochs but still very good accuracy
```
