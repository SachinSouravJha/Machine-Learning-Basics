```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(10)
print(x)
y=(x-5)**2
print(y)
```

```python
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("f(x)")
plt.show()
```

```python
x1=0
lr=0.1
error = []
plt.plot(x,y)
for i in range(50):
    grad=2*(x1-5)
    x1=x1-(lr*grad)
    y1=(x1-5)**2
    error.append(y1)
    plt.scatter(x1,y1)
    print(x1)
```

```python
plt.plot(error)
```
