```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(10)
y= x**2
print(x)
print(y)
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-dark')
z=2*x+2
plt.plot(x,y,color='red',label='parabola')
plt.plot(x,z,color='green',label='line')
plt.show()
```

```python
plt.style.use('seaborn-dark')
z=2*x+2
plt.scatter(x,y,color='red',label="parabola",marker='^')
plt.scatter(x,z,color='green',label="line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
x=np.array([0,1,2])*2
plt.bar(x,[100,10,30],width=0.5,color="red",tick_label=["CT","SK","SC"],label="Previous Year")
plt.bar(x+0.5,[10,30,80],width=0.5,color="green",label="Current Year")
plt.xlabel("Website")
plt.ylabel("Number of Artcles")
plt.legend()
plt.show()
```
