```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles,make_moons
#Importing two datasets
```

```python

X,Y = make_moons(n_samples=200,noise=0.10)

plt.scatter(X[:,0],X[:,1],c=Y)
plt.show()

#As we can see the two datasets are not lineraly separable
#So K means will perform badly here
```

```python

from sklearn.cluster import KMeans
km = KMeans(n_clusters=2)
km.fit(X)
#We only need two clusters here
```

```python
centers = km.cluster_centers_ #We took the two centers
label = km.labels_ #Taking all the labels

plt.scatter(X[:,0],X[:,1],c=label) #It will be coloured according to labels
plt.scatter(centers[:,0],centers[:,1],color='red') 
plt.show()
```

```python
#IN such cases DBSCAN is much better than KMeans
#In DBScan the algorithm tries to find high density regions with low density regions around it
#On the other hand K Means keeps trying to find high density points
```

```python
from sklearn.cluster import DBSCAN
db = DBSCAN(eps=0.3,min_samples=5) #For a low number it will give a lot of clusters
#min_samples defines that there should be at least five samples close to the point
#eps tells the distance
```

```python
db.fit(X)
```

```python
pred = db.fit_predict(X)
print(pred)
```

```python
#The problem with this is that we can't specify the number of clusters we want
plt.scatter(X[:,0],X[:,1],c=pred)
plt.show()
```
