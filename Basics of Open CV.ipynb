```python
import cv2
```

```python
img = cv2.imread('mario.png') #Normal pic

newImg = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)#Good quality pic because converted BGR into RGB
```

```python
from matplotlib import pyplot as plt
plt.imshow(img)
plt.show()
plt.imshow(newImg)
plt.show()
```

```python
import cv2

img = cv2.imread('mario.png')
gray = cv2.imread('mario.png',cv2.IMREAD_GRAYSCALE)
#Here we are using imshow method of open Cv instead of matplotlib because matplotlib kinda destroyed images by BGR conversion
cv2.imshow('Mario Image',img) #Normal Image
cv2.imshow('Grey Mario Image',gray) #Gray Image

cv2.waitKey(0) # Program will stop when any key is pressed
#If we pass some natural number the pic will stay for those number of seconds and then leave
cv2.destroyAllWindows()
```
