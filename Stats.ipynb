```
import pandas as pd
stats = pd.read_csv('Year_wise_batting_stats_all_ODI_players(20-12-19).csv?dl=0')
```

```
stats1 = stats[stats['Year']!="Overall"]
stats1.head()
```

```
stats1["Avg"] = stats1["Avg"].fillna(0)
stats1["S/R"] = stats1["S/R"].fillna(0)
stats1.head()
```

```
Unique = stats.player_name.unique()
Unique
```

```
stats1["Avg"] = stats1["Avg"]-stats["NO"]
stats1.head()
```

```
stats2 = stats1[stats1['Inns']>10]
stats2.head()
```

```
scaled_features = stats2.copy()
```

```
from sklearn.preprocessing import MinMaxScaler
col_names = ['Avg', 'S/R']
features = scaled_features[col_names]
scaler = MinMaxScaler().fit(features.values)
features = scaler.transform(features.values)
```

```
stats2.reset_index(drop=True, inplace=True)
stats2.head()
```

```
stats2.shape
```

```
impact = {}
experience = 1
for i in range(2440): 
  career = (stats2['player_name']==stats2.loc[i, "player_name"]).sum()
  if(career<4):
    experience = 0
  elif(career<8):
    experience = 0.6
  elif(career<12):
    experience = 0.8
  elif(career<16):
    experience = 1.0
  elif(career<16):
    experience = 1.2
  else:
    experience = 1.4
  if stats2.loc[i, "player_name"] in impact.keys(): 
    impact[stats2.loc[i, "player_name"]] += ((stats2.loc[i, "Avg"]*((stats2.loc[i, "S/R"])/3))/career)*experience
  else:
    impact[stats2.loc[i, "player_name"]] = ((stats2.loc[i, "Avg"]*((stats2.loc[i, "S/R"]/3)))/career)*experience
print(impact)
  
```

```
data_items = impact.items()
data_list = list(data_items)

df = pd.DataFrame(data_list,columns = ["Name","Points"])
print(df)
```

```
df.sort_values("Points", axis = 0, ascending = False, 
                 inplace = True)
print(df)
```
